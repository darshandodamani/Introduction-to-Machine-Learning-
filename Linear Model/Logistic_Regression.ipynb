```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([0, 0, 1, 1, 1])  # Binary classes
```

```python
# Plotting
plt.scatter(X, Y, color='blue', label='Data points')

# Logistic regression decision boundary (sigmoid function)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x_values = np.linspace(0, 6, 100)
y_values = sigmoid(2 * x_values - 5)  # Example coefficients

plt.plot(x_values, y_values, color='red', label='Logistic Regression Decision Boundary (Threshold = 0.5)')
plt.xlabel('X')
plt.ylabel('Probability')
plt.legend()
plt.title('Logistic Regression')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
X = np.array([1, 2, 3, 4, 5])
Y_linear = 2 * X + 1  # Linear relationship
Y_logistic = np.array([0, 0, 1, 1, 1])  # Binary classes

# Plotting
plt.figure(figsize=(10, 5))

# Linear regression
plt.subplot(1, 2, 1)
plt.scatter(X, Y_linear, color='blue', label='Data points')
plt.plot(X, 2 * X + 1, color='red', label='Linear Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Linear Regression')

# Logistic regression
plt.subplot(1, 2, 2)
plt.scatter(X, Y_logistic, color='blue', label='Data points')

# Logistic regression decision boundary (sigmoid function)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x_values = np.linspace(0, 6, 100)
y_values = sigmoid(2 * x_values - 5)  # Example coefficients

plt.plot(x_values, y_values, color='red', label='Logistic Regression  (thr = 0.5)')
plt.xlabel('X')
plt.ylabel('Probability')
plt.legend()
plt.title('Logistic Regression')

plt.tight_layout()
plt.show()
```
