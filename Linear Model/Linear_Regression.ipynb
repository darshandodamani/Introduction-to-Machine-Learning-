```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data
X = np.array([1, 2, 3, 4, 5])
Y = 2 * X + 1  # Linear relationship
```

```python

# Plotting
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, 2 * X + 1, color='red', label='Linear Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Linear Regression')
plt.show()
```
