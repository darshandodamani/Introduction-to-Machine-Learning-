```python
import numpy as np
import matplotlib.pyplot as plt

# Logistic loss function
def logistic_loss(x):
    return np.log(1 + np.exp(x))

# 0/1 loss function
def zero_one_loss(x):
    return np.where(x < 0, 1, 0)

# Hyperplane distance values
hyperplane_distance = np.linspace(-5, 5, 100)

# Compute losses
logistic_losses = logistic_loss(hyperplane_distance)
zero_one_losses = zero_one_loss(hyperplane_distance)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(hyperplane_distance, logistic_losses, label='Logistic Loss', color='blue')
plt.plot(hyperplane_distance, zero_one_losses, label='0/1 Loss', color='red')
plt.xlabel('Hyperplane Distance')
plt.ylabel('Loss')
plt.title('Loss over Hyperplane Distance')
plt.legend()
plt.grid(True)
plt.show()
```
