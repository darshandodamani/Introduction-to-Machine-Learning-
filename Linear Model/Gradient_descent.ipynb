```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Given data
examples = np.array([[4, 1], [1, -1], [5, 1]])
w = np.array([1, 2])


```

```python
# Plotting the line defined by w
x_line = np.linspace(0, 6, 100)
y_line = (-w[0] / w[1]) * x_line
plt.plot(x_line, y_line, label='Line defined by $\\mathbf{w}$')

# Plotting the examples from D
plt.scatter(examples[:, 0], examples[:, 1], marker='o', label='Examples from $D$')

# Highlighting the randomly selected example x1
plt.scatter(examples[0, 0], examples[0, 1], marker='x', color='red', label='Selected example $x_1$')

# Setting labels and title
plt.xlabel('$x_1$')
plt.ylabel('$c$')
plt.title('Line defined by $\\mathbf{w}$ and Examples from $D$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Show the plot
plt.show()
```
