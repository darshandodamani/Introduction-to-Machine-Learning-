```python
import matplotlib.pyplot as plt
import numpy as np

# Coordinates
x = [0, 0, 1, 1]
y = [0, 1, 0, 1]
colors = [0, 0, 1, 0]  # Decision boundary is 1 only for (1, 0)

# Perceptron weights and bias
w0, w1, w2 = -0.5, 0.5, 0.5

# Perceptron function
def perceptron(x1, x2):
    return np.heaviside(w0 + w1 * x1 + w2 * x2, 1)

# Calculate decision boundary line
x_line = np.linspace(-0.5, 1.5, 100)
y_line = (-w0 - w1 * x_line) / w2

# Plot points
plt.scatter(x, y, c=colors, marker='o', cmap='viridis')

# Plot decision boundary line
plt.plot(x_line, y_line, linestyle='--', color='red', label='Decision Boundary')

# Annotate points
for i, txt in enumerate(colors):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i] + 0.1, y[i] - 0.1))

# Annotate weights and bias
plt.annotate(f'w0 = {w0}', (-0.5, 1))
plt.annotate(f'w1 = {w1}', (-0.5, 0.8))
plt.annotate(f'w2 = {w2}', (-0.5, 0.6))

# Label axes
plt.xlabel('x1')
plt.ylabel('x2')

# Add title and legend
plt.title('Perceptron Graph with Decision Boundary')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step function to mimic the perceptron output
def step_function(x):
    return 1 if x >= 0 else 0

# Perceptron function
def perceptron(x, weights):
    return step_function(np.dot(x, weights[1:]) + weights[0])

# Perceptron training algorithm
def perceptron_training(x, y, weights, learning_rate):
    for i in range(len(x)):
        prediction = perceptron(x[i], weights)
        error = y[i] - prediction
        weights[1:] += learning_rate * error * x[i]
        weights[0] += learning_rate * error
    return weights

# Plotting the decision boundary
def plot_decision_boundary(weights):
    x_vals = np.linspace(-0.5, 1.5, 100)
    y_vals = (-weights[0] - weights[1] * x_vals) / weights[2]
    plt.plot(x_vals, y_vals, label='Decision Boundary')

# Training data
training_data = np.array([
    (0, 0, 0),
    (0, 1, 0),
    (1, 0, 1),
    (1, 1, 0)
])

# Initial weights
weights = np.array([-0.5, 0.5, 0.5])

# Plotting the initial decision boundary
plt.scatter(training_data[:, 0], training_data[:, 1], c=training_data[:, 2], cmap='viridis')
plot_decision_boundary(weights)
plt.title('Initial Decision Boundary')
plt.show()

# Training the perceptron
learning_rate = 0.3
for epoch in range(4):
    weights = perceptron_training(training_data[:, :2], training_data[:, 2], weights, learning_rate)
    
    # Plotting decision boundary after each update
    plt.scatter(training_data[:, 0], training_data[:, 1], c=training_data[:, 2], cmap='viridis')
    plot_decision_boundary(weights)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title(f'Decision Boundary after Epoch {epoch + 1}')
    plt.show()
```
